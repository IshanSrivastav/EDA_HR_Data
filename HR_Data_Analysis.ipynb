#Distribution of Each Variable

```python
for column in hr_df.columns:
  plt.figure(figsize=(10,5))
  sns.histplot(hr_df[column],kde=True, color="red")
  plt.title(f'Distribution of {column}')
  plt.show()
```

#Distibution of Salary level by department

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Department',hue='salary',data=hr_df,palette='Set1')
plt.title('Distribution of Salary Level By Department')
plt.show()
```

# Satisfaction Vs. Last Evaluation

```python
sns.scatterplot(x='satisfaction_level', y='last_evaluation', data=hr_df, color='steelblue')
plt.title('Satisfaction Vs Last Evaluation')
plt.show()
```

#Impact of each variable on employees leaving

```python
numerical_columns=['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 'time_spend_company']
for column in numerical_columns:
  sns.boxplot(x='left', y=column, data=hr_df, hue='left', palette='Set1', dodge=False)
  plt.title(f'Impact of {column} on left')
  plt.xlabel('Left(0=Stayed,1=Left)')
  plt.ylabel(column)
  plt.show()
```

```python
categorical_columns=['Department', 'salary', 'promotion_last_5years', 'Work_accident']
for column in categorical_columns:
  plt.figure(figsize=(10,5))
  sns.countplot(x=column,hue='left', data=hr_df, palette='Set1')
  plt.title(f'Impact of {column} on left')
  plt.xlabel(column)
  plt.ylabel('count')
  plt.show()
```

#Number of Project by salary

```python
plt.figure(figsize=(10,5))
sns.violinplot(x='salary',y='number_project', data=hr_df, hue='salary', palette='Set1')
plt.title('Number Of Project By Salary')
plt.xlabel('Salary')
plt.ylabel('Number Of Projects')
plt.show()
```

#Average Monthly hours by number of projects and salary level

```python
plt.figure(figsize=(10,5))
sns.catplot(x='number_project', y='average_montly_hours', hue='salary', data=hr_df, palette='Set1')
plt.grid()
plt.show()
```

#Correlation

```python
print(hr_df.dtypes)
```

```python
salary_mapping={'low':1,'medium':2,'high':3}
hr_df.loc[:,'salary_numeric']=hr_df['salary'].map(salary_mapping)
```

```python
print(hr_df.dtypes)
```

```python
departments=hr_df['Department'].unique()

for department in departments:
    dept_data = hr_df[hr_df['Department'] == department].select_dtypes(include='number')
    correlation_matrix = dept_data.corr()
    plt.figure(figsize=(10,8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title(f'Correlation Matrix for {department} Department')
    plt.show()
```
